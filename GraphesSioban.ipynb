```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Figure 1 : Eccentricite - Period distribution
```

```python
cols = ['pl_name','discoverymethod','pl_orbper','pl_orbeccen']
name = ['Planet name','Discovery method','Orbital Period [days]','Eccentricity']
df1 = pd.read_csv('PS_2023.10.31_06.07.23.csv', skiprows = 96, usecols = cols)
```

```python
discMeth = df1['discoverymethod'].unique()
#len(discMeth) = 11
# discMeth = array(['Radial Velocity', 'Imaging', 'Eclipse Timing Variations','Transit', 'Astrometry', 'Disk Kinematics', 'Microlensing','Orbital Brightness Modulation', 'Pulsation Timing Variations','Transit Timing Variations', 'Pulsar Timing'], dtype=object)
```

```python
couleurs = ['r','g','m','b','y','gray','c','m','k','chartreuse','brown']

# Initialisaion de la figure
groupe = df1[df1['discoverymethod'] == discMeth[0]]
figure1 = groupe.plot.scatter(x="pl_orbper", y="pl_orbeccen", logx=True,label=discMeth[0],ec=couleurs[0],fc='none')

for i in range(1,len(discMeth)) :
    methode = discMeth[i]
    groupe = df1[df1['discoverymethod'] == methode]
    figure1 = groupe.plot.scatter(x="pl_orbper", y="pl_orbeccen", logx=True,label=methode,ec=couleurs[i],fc='none', ax=figure1)
    
figure1.set_title("Eccentricity - Period distribution")
figure1.set_xlabel('Period [days]')
figure1.set_ylabel('Eccentricity')

figure1
```

```python
# Figure 2 : Irradiation - period distribution
cols2 = ['pl_name','discoverymethod','pl_orbper','pl_insol']
# pl_insol : Insolation Flux [Earth Flux]
df2 = pd.read_csv('PS_2023.10.31_06.07.23.csv', skiprows = 96, usecols = cols2)

discMeth = df2['discoverymethod'].unique()

couleurs = ['r','g','m','b','y','gray','c','m','k','chartreuse','brown']

# Initialisation de la figure
groupe = df2[df2['discoverymethod'] == discMeth[0]]
figure2 = groupe.plot.scatter(x="pl_orbper", y="pl_insol", logx=True,logy=True, label=discMeth[0], ec=couleurs[0],fc='none')

for i in range(1,len(discMeth)) :
    methode = discMeth[i]
    groupe = df2[df2['discoverymethod'] == methode]
    figure2 = groupe.plot.scatter(x="pl_orbper", y="pl_insol", logx=True,logy=True, label=methode, ec=couleurs[i],fc='none', ax=figure2)
    
figure2.set_title("Irradiation - Period distribution")
figure2.set_xlabel('Period [days]')
figure2.set_ylabel('Irradiation relative to Earth')

figure2
```
