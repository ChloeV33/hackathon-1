```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("table_planetes.csv", skiprows = 96)
df.dtypes
```

```python
"Orbital days = pl_orbper"
"Earth Radius = pl_rade"
"Mass Jupiter = pl_bmassj"
```

```python
by_discovery = df.groupby(by='discoverymethod')
df3 = by_discovery
discov = df['discoverymethod'].unique()
df1 = df[['pl_orbper', 'pl_bmassj']]
df2 = df[['pl_orbper', 'pl_rade']]
df.plot.scatter(x='pl_orbper', y = 'pl_bmassj',logy = True, logx = True)
discov
```

```python
discov = df['discoverymethod'].unique()
colors = ['Red', 'Blue', 'Yellow', 'Green', 'Gray', 'Black', 'Purple', 'Orange', 'Light Blue', 'Dark Blue', 'Dark Green']
extern = ['none', 'Blue', 'Yellow', 'none', 'Gray', 'none', 'Purple', 'Orange', 'none', 'none', 'none', 'none', 'none']
shapes = ["o", "*", "p", "s", "o", "s", "^", "s", ".", ".", ".", "."]
df4 = df3.get_group('Radial Velocity')
axr = df4.plot.scatter(x='pl_orbper', y = 'pl_bmassj',logy = True, logx = True, fc = 'none', ec = 'Red', label = 'Radial Velocity', marker =shapes[i])
for i in range(1,len(discov)):
    df4 = df3.get_group(discov[i])
    axr=df4.plot.scatter(x='pl_orbper', y = 'pl_bmassj',logy = True, logx = True, fc = extern[i], ec = colors[i], label = discov[i], ax = axr, marker = shapes[i], xlabel = 'Period (days)', ylabel = 'Mass (Jupiter Mass)')
axr
```
